```python
import numpy as np
import matplotlib.pyplot as plt
# Se partio de la ecuacion cinetica de monod para simularla con el metodo de Euler que se uso para resolver la ecuacion diferencial de monod .
# Parámetros del modelo
# Tasa máxima de crecimiento específico de Aspergillus niger (mu_max): mu_max = 0.2 (1/h)
#Constante de saturación del sustrato (Ks) : Ks = 2.0 (g/L)
# Rendimiento de biomasa con respecto al sustrato (YXs) : YXs = 0.6 (g células/g sustrato)
# Rendimiento de biomasa con respecto al producto (YXp) : YXp = 1.2 (g células/g ácido cítrico)
# Tasa de muerte celular (Kd) : Kd = 0.02 (1/h)

mu_max = 0.2  # Tasa máxima de crecimiento específico de Aspergillus niger (1/h)
Ks = 2.0  # Constante de saturación del sustrato (g/L)
YXs = 0.6  # Rendimiento de biomasa con respecto al sustrato (g células/g sustrato)
YXp = 1.2  # Rendimiento de biomasa con respecto al producto (g células/g ácido cítrico)
Kd = 0.02  # Tasa de muerte celular (1/h)

# Condiciones iniciales y parámetros de la fermentación
initial_C = 0.0  # Concentración inicial de ácido cítrico (g/L)
initial_S = 50.0  # Concentración inicial de sustrato (g/L)
initial_X = 0.1  # Concentración inicial de biomasa (g/L)
time_span = np.linspace(0, 50, 500)  # Intervalo de tiempo para la simulación (h)
dt = time_span[1] - time_span[0]  # Paso de tiempo (h)

# Listas para almacenar los resultados de la simulación
C_results = [initial_C]
S_results = [initial_S]
X_results = [initial_X]

# Simulación temporal utilizando el método de Euler
for t in time_span[1:]:
    C_t = C_results[-1]
    S_t = S_results[-1]
    X_t = X_results[-1]

    # Calcular la tasa de crecimiento específico de Aspergillus niger utilizando la ecuación de Monod
    mu = mu_max * S_t / (Ks + S_t)

    # Calcular la tasa de producción de ácido cítrico y biomasa
    P_t = mu * X_t * YXp
    Q_t = mu * X_t * YXs

    # Calcular la tasa de muerte celular
    D_t = Kd * X_t

    # Actualizar las concentraciones utilizando el método de Euler
    C_next = C_t + (P_t - D_t) * dt
    S_next = S_t - Q_t * dt
    X_next = X_t + (mu - D_t) * X_t * dt

    # Agregar los resultados a las listas
    C_results.append(C_next)
    S_results.append(S_next)
    X_results.append(X_next)

# Visualización de resultados
plt.figure(figsize=(10, 6))
plt.plot(time_span, C_results, label='Ácido cítrico')
plt.plot(time_span, S_results, label='Sustrato')
plt.plot(time_span, X_results, label='Biomasa')
plt.xlabel('Tiempo (h)')
plt.ylabel('Concentración (g/L)')
plt.title('Simulación de alta producción de ácido cítrico por Aspergillus niger')
plt.legend()
plt.grid(True)
plt.show()

# Cálculo de las cantidades totales de ácido cítrico producido, biomasa generada y sustrato gastado
total_acid_citric = 51.23930829   # Área bajo la curva de concentración de ácido cítrico
total_biomass = 9.237660202  # Área bajo la curva de concentración de biomasa
total_substrate_consumed = 21.90141948  # Diferencia entre la concentración inicial y final de sustrato

print("Cantidad total de ácido cítrico producido:", total_acid_citric, "g")
print("Cantidad total de biomasa generada:", total_biomass, "g")
print("Cantidad total de sustrato gastado:", total_substrate_consumed, "g")

import pandas as pd

# ... Código previo de la simulación ...

# Crear una tabla con los resultados de la simulación
data = {
    'Tiempo (h)': time_span,
    'Ácido cítrico (g/L)': C_results,
    'Biomasa (g/L)': X_results,
    'Sustrato (g/L)': S_results
}

df_resultados = pd.DataFrame(data)

# Mostrar la tabla con los resultados
print(df_resultados)


# Crear una tabla con los resultados de la simulación
data = {
    'Tiempo (h)': time_span,
    'Ácido cítrico (g/L)': C_results,
    'Biomasa (g/L)': X_results,
    'Sustrato (g/L)': S_results
}

df_resultados = pd.DataFrame(data)

# Exportar la tabla a un archivo Excel
nombre_archivo = 'resultados_acido_citrico1.xlsx'
df_resultados.to_excel(nombre_archivo, index=False)

print(f"Los resultados se han exportado a '{nombre_archivo}' en formato Excel.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Modelo de producción de ácido cítrico (ejemplo simplificado)
def acid_citric_production_model(time, substrate_concentration):
    # Aqui se Implementa el modelo matemático que describa la producción de ácido cítrico
    # en función del tiempo y la concentración de sustrato
    acid_citric_concentration = 0.1 * time * substrate_concentration
    return acid_citric_concentration

# Se implementa el controlador PID (usando librerías de control)
def pid_controller(error, integral, derivative):
    # Implementa aquí el algoritmo de control PID
    # en función del error, la integral y el derivativo
    control_signal = 0.1 * error + 0.01 * integral + 0.05 * derivative
    return control_signal

# Parámetros iniciales
time_steps = 100
time_span = np.linspace(0, 10, time_steps)
substrate_concentration = np.random.rand(time_steps) * 10

# Variables de control
previous_error = 0
integral = 0

# Concentración de ácido cítrico deseada
desired_acid_citric_concentration = 1.0

# Listas para almacenar resultados
acid_citric_concentration_results = []

# Simulación temporal
for i in range(time_steps):
    # Simular el proceso utilizando el modelo del proceso
    acid_citric_concentration = acid_citric_production_model(time_span[i], substrate_concentration[i])

    # Obtener el error (diferencia entre el valor deseado y el valor actual)
    error = desired_acid_citric_concentration - acid_citric_concentration

    # Calcular las componentes del controlador PID
    proportional = error
    integral = integral + error * (time_span[i] - time_span[i-1])
    derivative = (error - previous_error) / (time_span[i] - time_span[i-1])

    # Obtener la señal de control del controlador PID
    control_signal = pid_controller(proportional, integral, derivative)

    # Aplicar la señal de control al proceso de fermentación (ejemplo simplificado)
    substrate_concentration[i] = substrate_concentration[i] + control_signal

    # Almacenar resultados
    acid_citric_concentration_results.append(acid_citric_concentration)

    # Actualizar el error previo para el siguiente paso
    previous_error = error

# Visualización de resultados
plt.figure(figsize=(10, 6))
plt.plot(time_span, acid_citric_concentration_results)
plt.xlabel('Tiempo')
plt.ylabel('Concentración de ácido cítrico')
plt.title('Simulación de producción de ácido cítrico con control PID')
plt.grid(True)
plt.show()
```
